## Problem Statement::

## Null Hypothesis:: 
    H0 : Gender and Student Motivation are independent
## Alternative Hypothesis::
    Ha : Gender and Student Motivation are NOT independent
    
## say, alpha or sinificance level = 0.05

## SOLUTION::

```python
## import relevant libraries::
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_excel("/home/ashish/projects/TestOfIndependence_chiSquared/acad.xlsx")
dataframe
```

```python
##Cross table between gender and student motivation
cross_table = pd.pivot_table(dataframe[['g','sm']],index='g',columns='sm',aggfunc=len)
cross_table
```

```python
## Perform chi_square test to check independence of these variable
from scipy.stats import chi2_contingency
```

```python
chi_square,p_value,dof,tbl = chi2_contingency(cross_table)
```

```python
from scipy.stats import chi2
```

```python
## this is our chi-square(table value)
chi2.ppf(0.95,2)  ## Since for 5% significance --> (1-0.05)
```

## Using the p-value approach of hypothesis testing::
    1. If p ≤ α, reject H0; 
    2. otherwise, if p > α do not reject H0.


    since, p-value=0.30 > α=0.05  ==> Do not reject H0.
    This implies
## Gender and Student motivation are independent.

## USing the critical value approach::
    1. If the value of the test statistic falls in the rejection region, reject H0; 
    2. otherwise, do not reject H0
    -------------------OR-----------------------------------------
    If the observed chi square test statistic is greater than critical value,the null  hypotheses can be rejected.
    
    since,chi2(calculated) = 5.991464547107979 < chi2(alpha,dof) = 0.10.
## Thus, Do not Reject H0. 
# This implies Gender and Student motivation are independent.
